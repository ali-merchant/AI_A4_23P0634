```python
# 0 1 2
# 3 4 5
# 6 7 8

terminal = [
    (0, 1, 2), (3, 4, 5), (6, 7, 8),  
    (0, 3, 6), (1, 4, 7), (2, 5, 8), 
    (0, 4, 8), (2, 4, 6)              
]
```

```python
def print_board(board):
    for i in range(0, 9, 3):
        row = ""
        for j in range(3):
            index = i + j
            if board[index] == " ":
                row += str(index)
            else:
                row += board[index]

            if j != 2:
                row += " | "
        print(" " + row)

        if i != 6:
            print("---+---+---")
```

```python
def check_terminal(board):
    for line in terminal:
        a = line[0]
        b = line[1]
        c = line[2]

        if board[a] != " ":
            if board[a] == board[b] and board[b] == board[c]:
                return board[a]

    return None
```

```python
def is_full(board):
    for cell in board:
        if cell == " ":
            return False
    return True
```

```python
def is_terminal(board):
    if check_terminal(board) is not None:
        return True

    if is_full(board):
        return True

    return False
```

```python
def get_utility(board):
    winner = check_terminal(board)

    if winner == "X":
        return 1
    elif winner == "O":
        return -1
    else:
        return 0
```

```python
def get_moves(board):
    moves = []

    for i in range(9):
        if board[i] == " ":
            moves.append(i)

    return moves

def make_move(board, move, player):
    new_board = board.copy()
    new_board[move] = player
    return new_board
```

```python
def min_value(board, alpha, beta):
    if is_terminal(board):
        return get_utility(board)

    value = 999

    moves = get_moves(board)

    for move in moves:
        new_board = make_move(board, move, "O")

        child_value = max_value(new_board, alpha, beta)

        if child_value < value:
            value = child_value

        # update beta
        if value < beta:
            beta = value

        # pruning
        if alpha >= beta:
            break

    return value



def max_value(board, alpha, beta):
    if is_terminal(board):
        return get_utility(board)

    value = -999

    moves = get_moves(board)

    for move in moves:
        new_board = make_move(board, move, "X")

        child_value = min_value(new_board, alpha, beta)

        if child_value > value:
            value = child_value

        # update alpha
        if value > alpha:
            alpha = value

        # pruning
        if alpha >= beta:
            break

    return value

```

```python
def choose_move(board, player):
    best_move = None

    if player == "X":
        best_value = -999

        for move in get_moves(board):
            new_board = make_move(board, move, "X")

            value = min_value(new_board, -999, 999)

            if value > best_value:
                best_value = value
                best_move = move

    else:
        best_value = 999

        for move in get_moves(board):
            new_board = make_move(board, move, "O")

            value = max_value(new_board, -999, 999)

            if value < best_value:
                best_value = value
                best_move = move

    return best_move
```

```python
board = [" "] * 9
player = "X"

print("Start Game:\n")
print_board(board)

while not is_terminal(board):
    print("\nPlayer", player, "is thinking...")

    move = choose_move(board, player)

    board = make_move(board, move, player)

    print("\nMove chosen:", move)
    print_board(board)

    if player == "X":
        player = "O"
    else:
        player = "X"

# Game result
winner = check_terminal(board)

if winner == "X":
    print("\nX wins!")
elif winner == "O":
    print("\nO wins!")
else:
    print("\nDraw!")
```
