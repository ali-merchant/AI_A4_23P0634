```python
# 0 1 2
# 3 4 5
# 6 7 8

terminal = [
    (0, 1, 2), (3, 4, 5), (6, 7, 8),  
    (0, 3, 6), (1, 4, 7), (2, 5, 8), 
    (0, 4, 8), (2, 4, 6)              
]
```

```python
def print_board(board):
    for i in range(0, 9, 3):
        row = ""
        for j in range(3):
            index = i + j
            if board[index] == " ":
                row += str(index)
            else:
                row += board[index]

            if j != 2:
                row += " | "
        print(" " + row)

        if i != 6:
            print("---+---+---")
```

```python
def check_terminal(board):
    for line in terminal:
        a = line[0]
        b = line[1]
        c = line[2]

        if board[a] != " ":
            if board[a] == board[b] and board[b] == board[c]:
                return board[a]

    return None
```

```python
def is_full(board):
    for cell in board:
        if cell == " ":
            return False
    return True
```

```python
def is_terminal(board):
    if check_terminal(board) is not None:
        return True

    if is_full(board):
        return True

    return False
```

```python
def get_utility(board):
    winner = check_terminal(board)

    if winner == "X":
        return 1
    elif winner == "O":
        return -1
    else:
        return 0
```

```python
def get_moves(board):
    moves = []

    for i in range(9):
        if board[i] == " ":
            moves.append(i)

    return moves
```
